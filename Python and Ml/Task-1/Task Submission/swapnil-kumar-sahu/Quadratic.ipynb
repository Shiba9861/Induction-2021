```python
import math

def equationroots( a, b, c):
	
	D = b * b - 4 * a * c
	sq = math.sqrt(abs(D))
	
	if D > 0:
		print(" Roots are real and different ")
		print((-b + sq)/(2 * a))
		print((-b - sq)/(2 * a))
	
	elif D == 0:
		print(" Roots are real and equal")
		print(-b / (2 * a))
	
	else:
		print("Complex Roots")
		print(- b / (2 * a), " + i", sq)
		print(- b / (2 * a), " - i", sq)

a = int(input("Enter coeff. a - "))
b = int(input("Enter coeff. b - "))
c = int(input("Enter coeff. c - "))

equationroots( a, b, c)
```
