```python
M = int(input("Enter 1st num- "))
sym = (input("Enter symbol "))
N = int(input("Enter 2nd num- "))
if sym == "+":
    print(M+N)
elif sym == "-":
    print(M-N)
elif sym == "*":
    print(M*N)
elif sym == "/":
    print(M/N)
elif sym == "%":
    print(M % N)
else:
    print("Invalid input !!")
```
