```python
def add(a, b):
    return a + b


# THE subtract FUNCTION TO PERFORM SUBTRACTION FOR TWO NUMBERS
def subtract(a, b):
    return a - b


# THE multiply FUNCTION TO PERFORM MULTIPLICATION OF TWO NUMBERS
def multiply(a, b):
    return a * b


# THE DIVIDE FUNCTION TO PEFORM DIVISION FOR TWO NUMBERS
def divide(a, b):
    return a / b


# DISPLAY THE MENU
print("CHOOSE AN OPERATION. ")
print("1. ADD")
print("2. SUBTRACT")
print("3. MULTIPLY")
print("4. DIVIDE")

# GET USER INPUT
operator = input("Select an operation (1,2,3,4): ")



number1 = int(input("Enter first number:"))
number2 = int(input("Enter second number:"))

# CONVERT THE INPUTS TO INTEGER
number1 = int(number1)
number2 = int(number2)

if operator == '1':
    result = add(number1, number2)
    print(number1, "+", number2, "=", result)

elif operator == '2':
    result = subtract(number1, number2)
    print(number1, "-", number2, "=", result)

elif operator == '3':
    result = multiply(number1, number2)
    print(number1, "*", number2, "=", result)

elif operator == '4':
    result = divide(number1, number2)
    print(number1, "/", number2, "=", result)

else:
    print("Invalid input")
```
