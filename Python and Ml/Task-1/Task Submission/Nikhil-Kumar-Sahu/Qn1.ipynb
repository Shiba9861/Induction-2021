```python
import math
a=int(input("a:"))
b=int(input("b:"))
c=int(input("c:"))
dis=b*b-4*a*c
sr=math.sqrt(abs(dis))
if dis>0:
    print("roots are distinct and real")
    print((-b+sr)/(2*a))
elif dis<0:
    print("roots are imaginary")
    print(-b/(2*a),"+ i",sr)
    print(-b/(2*a),"- i",sr)
else: 
    print("roots are equal")
    print(-b/(2*a))
```
