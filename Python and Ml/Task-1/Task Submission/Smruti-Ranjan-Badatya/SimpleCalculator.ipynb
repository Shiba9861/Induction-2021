```python
a = int(input("Enter Your First No."))
b = int(input("Enter Your Second Number No."))

c = input("Enter Your Operation( + - * / // ** % )")

if c == "+":
    print(a+b)
elif c == "-":
    print(a-b)
elif c == "*":
    print(a*b)
elif c == "/":
    print(a/b)
elif c == "//":
    print(a//b)
elif c == "**":
    print(a**b)
elif c == "%":
    print(a%b)
```
