```python
import math

a = float(input())
b = float(input())
c = float(input())

D = b**2 - 4*a*c

if D>=0:
    r1 = -b/2 - math.sqrt(D)/2
    r2 = -b/2 + math.sqrt(D)/2
else:
    r = -b/2
    i = math.sqrt(-D)

if D > 0:
    print("Real and Unique Roots and the roots are {0} {1}".format(r1,r2))
elif D == 0:
    print("Real and Equal Roots and the root is {0}".format(r1))
else:
    print("Imaginary roots and the roots are {0} + i{1} and {0} - i{1}".format(r,i))
```
