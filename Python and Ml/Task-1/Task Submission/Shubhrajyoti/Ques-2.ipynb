```python
#Program to find an Armstrong number within a given interval  
```

```python
lower = int(input("Enter lower value : "))
upper = int(input("Enter upper value : "))
print("Armstrong number in the given interval is : ")

for num in range(lower, upper + 1):
   power = len(str(num))
   sum=0
   temp=num
   while temp > 0:
       rem = temp % 10
       sum +=rem **power
       temp//= 10
   if num == sum:
       print(num)
```
