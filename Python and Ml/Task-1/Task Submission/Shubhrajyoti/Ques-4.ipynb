```python
#Program to sort words in a string in alphabetical order
```

```python
my_str = input("Enter a string : ")

words= my_str.split()               #split() function is used for splitting the whole string into separate words in a list
words.sort()                        #sort() function is used for sorting alphabetically all the list items(i.e. the words of string)  

print("The sorted words in alphabetical order are :")
for w in words:
   print(w)
```
