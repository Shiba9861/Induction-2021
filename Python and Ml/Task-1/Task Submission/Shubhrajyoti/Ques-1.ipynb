```python
#Program to solve the quadratic equation ax**2+bx+c=0
```

```python
import cmath
a=float(input("Enter value of a : "))
b=float(input("Enter value of b : "))
c=float(input("Enter value of c : "))
if a==0:
  print("Roots can't be found! Input correct value of a...")
else:
  d=b**2-4*a*c
  sqrt_d=cmath.sqrt(d)
  x1=(-b+sqrt_d)/(2*a)
  x2=(-b-sqrt_d)/(2*a)
  print("The roots of given quadratic equation are :{0} and {1}. " .format(x1,x2))
```
