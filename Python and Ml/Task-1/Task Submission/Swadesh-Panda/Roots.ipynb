```python
import math
print('Enter coefficients a,b,c in ax^2+bx+c=0 :')
a=float(input('a:'))
b=float(input('b:'))
c=float(input('c:'))
d=(b*b-4*a*c)
if d>0:
    print('Roots are:')
    print((-b+math.sqrt(d))/2*a)
    print((-b-math.sqrt(d))/2*a)
elif d==0:
    print('Both the roots are equal.')
    print(-b/2*a)
else:
    print('Roots are imagenary.')
```
